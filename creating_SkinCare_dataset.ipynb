 ### Sentiment analysis on skin care products

##  1. Download data from [Amazon product data](https://jmcauley.ucsd.edu/data/amazon/)

*We need "reviews_Beauty.json.gz" and "meta_Beauty.json.gz" files.*

##  2. Read the data into pandas data frame

#### 2.1 Reading the review dataset into a data frame

```python
import pandas as pd
import gzip
import json

def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield json.loads(l)

def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')

df_reviews = getDF("reviews_Beauty.json.gz")
```

```python
df_reviews
```

*The data has 2023070 entries and 9 columns:*

- reviewerID - ID of the reviewer
- asin - ID of the product
- reviewerName - name of the reviewer 
- helpful - helpfulness rating of the review
- reviewText - text of the review 
- overall - rating of the product 
- summary - summary of the review 
- unixReviewTime - time of the review (unix time) 
- reviewTime - time of the review (raw)

#### 2.2 Reading the meta data into a data frame

```python
def parseMeta(path): 
    g = gzip.open(path, 'r')
    for l in g:
        yield json.dumps(eval(l))

f = open("output.strict", 'w')
for l in parseMeta("meta_Beauty.json.gz"):
    f.write(l + '\n')
```

```python
def getDF_meta(path):
    df = pd.read_json(path, lines=True)
    return df

df_meta = getDF_meta("output.strict")
```

```python
df_meta
```

*Meta data includes product description, title, urls, salesRank, categories, price, related products and brand.*



##  3. Create skin care dataset

#### 3.1 Choosing skin care products

*Since we only interested in 'Skin Care' products, we select the products that includes 'Skin Care' in their categories.*   

```python
selected_ind = []
for i in range(len(df_meta)):
    if 'Skin Care' in df_meta.categories[i][0]:
        selected_ind.append(i)
```

```python
df_skin_care = df_meta.iloc[selected_ind]
```
